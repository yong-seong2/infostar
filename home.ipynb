# 인기 대외활동 크롤링 (캠퍼스픽)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# ChromeDriver 경로 설정
chrome_driver_path = "C:\\Users\\최라윤\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_experimental_option('detach', True)
chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])

# ChromeDriver 서비스 설정
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# CampusPick 인기 대외활동 페이지 열기
driver.get("https://www.campuspick.com/activity")

# WebDriverWait을 통해 요소가 로드될 때까지 대기
wait = WebDriverWait(driver, 10)

# 인기 대외활동 크롤링
activities = []

try:
    # 5개의 대외활동 정보 추출
    for i in range(5):
        try:
            # activity_items를 반복마다 다시 가져오기
            activity_items = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a.item")))

            # 각 대외활동 정보 추출
            name = activity_items[i].find_element(By.TAG_NAME, "h3").text
            poster = activity_items[i].find_element(By.TAG_NAME, "figure").get_attribute("data-image")
            dday = activity_items[i].find_element(By.CLASS_NAME, "dday").text
            viewcount_text = activity_items[i].find_element(By.CLASS_NAME, "viewcount").text

            # 조회수(view_count) 정수로 변환, 변환 실패 시 기본값 0 설정
            view_count = int(viewcount_text.replace(",", "").strip()) if viewcount_text.replace(",", "").isdigit() else 0

            activities.append((name, poster, dday, view_count))
        except Exception as e:
            print(f"오류 발생: {e}")
            time.sleep(1)  # 에러 발생 시 잠시 대기

finally:
    # 브라우저 종료
    driver.quit()

# 크롤링된 데이터 출력
for activity in activities:
    print(f"Name: {activity[0]}\nPoster URL: {activity[1]}\nD-Day: {activity[2]}\nView Count: {activity[3]}\n")
```

# 마감임박 공모전 및 대외활동 (위비티)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# ChromeDriver 경로 설정
chrome_driver_path = "C:\\Users\\최라윤\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_experimental_option('detach', True)
chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])

# ChromeDriver 서비스 설정
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Wevity 공모전 페이지 열기
driver.get("https://www.wevity.com/?c=find&s=1&gub=&cidx=&sp=&sw=&gbn=list&mode=soon")

# WebDriverWait을 통해 요소가 로드될 때까지 대기
wait = WebDriverWait(driver, 10)

try:
    # 공모전 리스트 추출 (모든 li 요소)
    contest_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#container > div.content-area > div.content-wrap > div.content > div:nth-child(4) > div > ul > li")))

    # 공모전 정보 추출
    for i, contest_element in enumerate(contest_elements[:5]):  # 최대 5개 추출
        try:
            # 공모전 이름과 D-Day 요소 추출
            contest_name = contest_element.find_element(By.CSS_SELECTOR, "div.tit > a").text
            dday = contest_element.find_element(By.CSS_SELECTOR, "div.day").text

            # 결과 출력
            print(f"공모전 {i + 1} 이름: {contest_name}")
            print(f"공모전 {i + 1} 남은 D-Day: {dday}")
        except Exception as e:
            print(f"오류 발생 (공모전 {i + 1}): {e}")
            continue

except Exception as e:
    print(f"오류 발생: {e}")

finally:
    # 브라우저 종료
    driver.quit()
```

# 한번에 크롤링 (인기 대외활동, 공모전, 마감임박)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

# ChromeDriver 경로 설정
chrome_driver_path = "C:\\Users\\최라윤\\Downloads\\chromedriver-win64\\chromedriver-win64\\chromedriver.exe"

# Chrome 옵션 설정
chrome_options = Options()
chrome_options.add_experimental_option('detach', True)
chrome_options.add_experimental_option('excludeSwitches', ['enable-logging'])

# ChromeDriver 서비스 설정
service = Service(chrome_driver_path)

# 결과 저장용 리스트
contests_data = []
activities_data = []
wevity_data = []

# WebDriverWait 설정
def initialize_driver():
    return webdriver.Chrome(service=service, options=chrome_options)

# CampusPick 공모전 크롤링
try:
    driver = initialize_driver()
    driver.get("https://www.campuspick.com/contest")
    wait = WebDriverWait(driver, 10)
    for i in range(5):
        try:
            contest_links = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a.item")))
            name = contest_links[i].find_element(By.TAG_NAME, "h3").text
            poster = contest_links[i].find_element(By.TAG_NAME, "figure").get_attribute("data-image")
            dday = contest_links[i].find_element(By.CLASS_NAME, "dday").text
            viewcount_text = contest_links[i].find_element(By.CLASS_NAME, "viewcount").text
            view_count = int(viewcount_text.replace(",", "").strip()) if viewcount_text.replace(",", "").isdigit() else 0
            contests_data.append(("CampusPick Contest", name, poster, dday, view_count))
        except Exception as e:
            print(f"CampusPick Contest 오류 발생: {e}")
finally:
    driver.quit()

# CampusPick 대외활동 크롤링
try:
    driver = initialize_driver()
    driver.get("https://www.campuspick.com/activity")
    wait = WebDriverWait(driver, 10)
    for i in range(5):
        try:
            activity_items = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "a.item")))
            name = activity_items[i].find_element(By.TAG_NAME, "h3").text
            poster = activity_items[i].find_element(By.TAG_NAME, "figure").get_attribute("data-image")
            dday = activity_items[i].find_element(By.CLASS_NAME, "dday").text
            viewcount_text = activity_items[i].find_element(By.CLASS_NAME, "viewcount").text
            view_count = int(viewcount_text.replace(",", "").strip()) if viewcount_text.replace(",", "").isdigit() else 0
            activities_data.append(("CampusPick Activity", name, poster, dday, view_count))
        except Exception as e:
            print(f"CampusPick Activity 오류 발생: {e}")
finally:
    driver.quit()

# Wevity 공모전 크롤링
try:
    driver = initialize_driver()
    driver.get("https://www.wevity.com/?c=find&s=1&gub=&cidx=&sp=&sw=&gbn=list&mode=soon")
    wait = WebDriverWait(driver, 10)
    contest_elements = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#container > div.content-area > div.content-wrap > div.content > div:nth-child(4) > div > ul > li")))
    for i, contest_element in enumerate(contest_elements[:5]):
        try:
            contest_name = contest_element.find_element(By.CSS_SELECTOR, "div.tit > a").text
            dday = contest_element.find_element(By.CSS_SELECTOR, "div.day").text
            wevity_data.append(("Wevity Contest", contest_name, "N/A", dday, "N/A"))
        except Exception as e:
            print(f"Wevity 오류 발생: {e}")
finally:
    driver.quit()

# 결과 병합
all_data = contests_data + activities_data + wevity_data
columns = ["Source", "Name", "Poster URL", "D-Day", "View Count"]
df = pd.DataFrame(all_data, columns=columns)

# 데이터 출력
print(df)
```
